```python
#create a team totals dataset to create a team average batting average per year based on the batters dataset
team_totals = BattingData.groupby(['yearID', 'teamID'])[['H', 'AB']].sum().reset_index()

team_totals['Team_Batting_Average'] = team_totals['H'] / team_totals['AB']

print(team_totals.head())
```

```python
#merged this back with the batting data
BattingData = BattingData.merge(team_totals[['yearID', 'teamID', 'Team_Batting_Average']], on=['yearID', 'teamID'], how='left')
```

```python
#filled in data with not hit by pitch as 0
TeamsData['HBP'] = TeamsData['HBP'].fillna(0)
```

```python
#calculation of team obp
TeamsData["Team_OBP"] = (TeamsData["H"]+TeamsData["BB"]+TeamsData["HBP"])/TeamsData["AB"]
TeamsData
```

```python
#create variable that has the total amount of singles within each team
TeamsData["1B"] = TeamsData["H"] - TeamsData["2B"] - TeamsData["3B"] - TeamsData["HR"]
```

```python
TeamsData[["H","1B","2B","3B","HR"]]
```

```python
#create team slugging percentage variable
TeamsData["Team_SLG"] = (TeamsData["1B"] + 2*TeamsData["2B"] + 3*TeamsData["3B"] + 4*TeamsData["HR"])/TeamsData["AB"]
```

```python
TeamsData
```

```python
#create team ops variable
TeamsData["Team_OPS"] = TeamsData["Team_OBP"] + TeamsData["Team_SLG"]
```

```python
#calculate total team payroll for each team
# SalaryData["Total_Team_Payroll"] = SalaryData.groupby(['yearID', 'teamID'])['salary'].sum().reset_index()
total_salaries = SalaryData.groupby(['yearID', 'teamID'])['salary'].sum().reset_index()
```

```python
#merge total team payroll to the teams data dataset
TeamsData = TeamsData.merge(total_salaries, on=['yearID', 'teamID'], how='left')
```
